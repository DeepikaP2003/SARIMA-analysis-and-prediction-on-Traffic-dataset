```python
#identifying if our data is stationary or not through mean and standard deviation as parameters.
df['rollMean']= df.Vehicles.rolling(window=12).mean()
df['rollStd']= df.Vehicles.rolling(window=12).std()

plt.figure(figsize=(10,5))
sns.lineplot(data=df,x=df['Date_no'],y=df['Vehicles'])
sns.lineplot(data=df,x=df['Date_no'],y=df.rollMean)
sns.lineplot(data=df,x=df['Date_no'],y=df.rollStd)
```

```python
#checking stationarity using ADF
from statsmodels.tsa.stattools import adfuller
adf_result = adfuller(df['Vehicles'],autolag='AIC')
adf_statistic = adf_result[0]
adf_p_value = adf_result[1]
adf_critical_values = adf_result[4]

print("ADF p-value:", adf_p_value)
print("ADF Statistic:", adf_statistic)
print("ADF Critical Values:", adf_critical_values)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Sample data creation (replace this with your actual data loading)
date_rng = pd.date_range(start='2015-01-01', end='2017-12-31', freq='D')
df = pd.DataFrame(date_rng, columns=['DateTime'])
df['Vehicles'] = np.random.randint(50, 150, size=(len(date_rng)))

# Split the data into training and testing sets
train_size = int(len(df) * 0.6)
train, test = df[:train_size], df[train_size:]

# SARIMA parameters (replace with your actual values)
p, d, q = 1, 1, 1
P, D, Q, s = 1, 1, 1, 24

# Fit SARIMA model on the training set
model = SARIMAX(train['Vehicles'], order=(p, d, q), seasonal_order=(P, D, Q, s))
results = model.fit()

# Make predictions on the testing set
forecast_steps = len(test)
forecast = results.get_forecast(steps=forecast_steps)

# Get the predicted values and confidence intervals
forecast_values = forecast.predicted_mean
forecast_conf_int = forecast.conf_int()

# Calculate Mean Squared Error (MSE) on the testing set

# Plot observed, predicted, and confidence intervals
plt.figure(figsize=(15, 6))
plt.plot(train.index, train['Vehicles'], label='Training', color='blue')
plt.plot(test.index, test['Vehicles'], label='Testing', color='green')
plt.plot(forecast_values.index, forecast_values.values, label='Forecast', color='red')
plt.fill_between(
    forecast_conf_int.index,
    forecast_conf_int.iloc[:, 0],
    forecast_conf_int.iloc[:, 1],
    color='grey',
    alpha=0.3,
)
plt.xlabel('Date and Time')
plt.ylabel('Vehicles')
plt.legend()
plt.title('SARIMA Forecast')
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Assuming you have true values (actual observations) and forecasted values
# Replace 'true_values' and 'forecast_values' with your actual variable names

# Mean Absolute Error (MAE)
mae = mean_absolute_error(true_values, forecast_values)

# Mean Squared Error (MSE)
mse = mean_squared_error(true_values, forecast_values)

# Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# Mean Absolute Percentage Error (MAPE)
mape = np.mean(np.abs((true_values - forecast_values) / true_values)) * 100

# Print the metrics
print(f'Mean Absolute Error (MAE): {mae}')
print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
print(f'Mean Absolute Percentage Error (MAPE): {mape}%')


```
